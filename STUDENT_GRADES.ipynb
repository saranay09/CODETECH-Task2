```python
"""
  Develop a Python program to track and manage student grades. The program should allow the user to input grades for different subjects or
assignments, calculate the average grade, and display the overall grade along with any additional information (such as letter grade or
GPA).

"""
def calculate_grade(grades):

  # Calculate the average grade
  average = sum(grades) / len(grades)

  # Assign letter grade based on a grading scale
  if average >= 90:
    letter_grade = "O"
  elif average >= 80:
    letter_grade = "A+"
  elif average >= 70:
    letter_grade = "A"
  elif average >= 60:
    letter_grade = "B+"
  elif average >= 50:
    letter_grade = "B"
  elif average >= 45:
    letter_grade = "C"
  else:
    letter_grade = "RA"

  return average, letter_grade

def main():
  """
  Prompts user to input grades, calculates average, and displays results.
  """

  grades = []
  num_grades = int(input("Enter the number of grades to input: "))

  student_name = input("Enter the name of student :")
    
  course = input("\nThe course of student is:")
  

  # Get user input for each grade
  for i in range(num_grades):
    while True:
      try:
        grade = float(input(f"Enter grade {i+1}: "))
        if 0 <= grade <= 100:
          grades.append(grade)
          break
        else:
          print("Invalid grade. Please enter a value between 0 and 100.")
      except ValueError:
        print("Invalid input. Please enter a number.")

  # Calculate average and letter grade
  average, letter_grade = calculate_grade(grades)

  # Display results
  print("\nName : ",student_name)
    
  print("\nCourse : ",course)

  print(f"\nAverage grade: {average:.2f}")
    
  print(f"\nGPA : {average/10:.2f}")  
  
  print(f"GRADE: {letter_grade}")

if __name__ == "__main__":
  main()
```
